```python
# Imports
import pandas as pd
from sqlalchemy import create_engine, text
import os
import urllib.parse
from dotenv import load_dotenv
```

```python
#Environment
# Load DB credentials from root .env
load_dotenv('../.env')  
```

```python
# Engine Creation
USER = os.getenv("MYSQL_USER")
PASS = urllib.parse.quote_plus(os.getenv("MYSQL_PASSWORD")) 
DB = os.getenv("MYSQL_DATABASE")
PORT = os.getenv("HOST_PORT", "3307")

engine = create_engine(f"mysql+pymysql://{USER}:{PASS}@127.0.0.1:{PORT}/{DB}")
```

```python
# Test the connection

try:
    with engine.connect() as connection:
        # A simple query that works on any MySQL database
        try:
            with engine.connect() as connection:
                result = connection.execute(text("SELECT 1"))
                print("✅ Successfully connected to the MySQL container!")
        except Exception as e:
            print(f"❌ Connection failed: {e}")
        print("✅ Successfully connected to the MySQL container!")
except Exception as e:
    print(f"❌ Connection failed: {e}")
```

```python
df = pd.read_sql("SELECT * FROM your_table", engine)
print(df.head())
```

```python
# Group expenses (debits) by payee (Description) and sum the Amount
df['amount'] = pd.to_numeric(df['amount'], errors='coerce')
expenses = df[df['amount'] < 0].dropna(subset=['amount'])
grouped_expenses = expenses.groupby('description')['amount'].sum().sort_values(ascending=False)
print(grouped_expenses)
grouped_expenses.head()
```

```python
# Group total income (credits) by payee (Description) and sum the Amount for each
grouped_income = df[df['Type'] == 'Credit'].groupby('Description')['Amount'].sum().sort_values(ascending=False)
print(grouped_income)
grouped_income.head()
```
